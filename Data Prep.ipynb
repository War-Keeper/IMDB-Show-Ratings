```python
# Data.tsv is stored locally in the 
# same directory as of this python file
basics = pd.read_csv('basics.tsv',sep = '\t')
basics_episodes = basics.loc[basics['titleType'] == "tvEpisode"]

basics_episodes
```

```python
basics_series = basics.loc[basics['titleType'] == "tvSeries"]

basics_series
```

```python
ratings = pd.read_csv('ratings.tsv',sep = '\t')
ratings
```

```python
episodes = pd.read_csv('episodes.tsv',sep = '\t')
episodes
```

```python
episode_ratings_original = pd.merge(ratings,episodes, left_on = 'tconst', right_on = 'tconst', how = 'inner')
episode_ratings = pd.merge(ratings,episodes, left_on = 'tconst', right_on = 'tconst', how = 'inner')

mask = episode_ratings['seasonNumber'] == '\\N'
episode_ratings = episode_ratings[~mask]
mask = episode_ratings['episodeNumber'] == '\\N'
episode_ratings = episode_ratings[~mask].reset_index()

episode_ratings
```

```python
basic_episode_ratings = pd.merge(episode_ratings,basics_episodes, left_on = 'tconst', right_on = 'tconst', how = 'inner')

basic_episode_ratings = basic_episode_ratings[['tconst', 'averageRating', 'parentTconst', 'seasonNumber', 'episodeNumber', 
                                               'titleType', 'primaryTitle']]
basic_episode_ratings
```

```python
# basic_series_ratings = pd.merge(episode_ratings_original,basics_series, left_on = 'parentTconst', right_on = 'tconst', how = 'inner')

# basic_series_ratings
```

```python
basic_series_ratings = pd.merge(ratings,basics_series, left_on = 'tconst', right_on = 'tconst', how = 'inner')

mask = basic_series_ratings['averageRating'] == '\\N'
basic_series_ratings = basic_series_ratings[~mask]

basic_series_ratings = basic_series_ratings[['tconst', 'averageRating', 'titleType', 'primaryTitle', 'originalTitle']]
basic_series_ratings
```
